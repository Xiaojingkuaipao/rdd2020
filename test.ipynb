```python
from PIL import Image
import torch
from torchvision import transforms as T


image_path = "./images/Japan_010313.jpg"
img = Image.open(image_path)
img.show()

img_tensor = T.ToTensor()(img)
img_tensor.shape
```

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
resized_640 = F.interpolate(img_tensor.unsqueeze(0), size=(640, 640), mode='bilinear', align_corners=False).squeeze(0)

# Interpolate to 1280x1280 
resized_1280 = F.interpolate(img_tensor.unsqueeze(0), size=(1280, 1280), mode='bilinear', align_corners=False).squeeze(0)

# Visualize original and resized images
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Original image
axes[0].imshow(img_tensor.permute(1, 2, 0))
axes[0].set_title('Original (600x600)')

# 640x640
axes[1].imshow(resized_640.permute(1, 2, 0))
axes[1].set_title('Resized 640x640')

# 1280x1280
axes[2].imshow(resized_1280.permute(1, 2, 0))
axes[2].set_title('Resized 1280x1280')
```

```python
img_1280 = T.ToPILImage()(resized_1280)
img_1280.show()
```
